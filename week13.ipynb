```python
import csv
import matplotlib.pyplot as plt

f = open('ta_20210527134343.csv')

data = csv.reader(f)

header = next(data)

max_temp = []
min_temp = []

for row in data:
    if row != []:
        month = row[0].split('-')[1]
        if month == '04':
            day = row[0].split('-')[2]
            if day == '07':
                max_temp.append(float(row[-1]))
                min_temp.append(float(row[-2]))

plt.plot(max_temp, 'r')
plt.plot(min_temp, 'b')
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('ta_20210527134343.csv')

data = csv.reader(f)

header = next(data)

jan = []
aug = []

for row in data:
    if row != []:
        month = row[0].split('-')[1]
        if month == '01':
            jan.append(float(row[-1]))
        if month == '08':
            aug.append(float(row[-1]))

plt.hist(jan, bins = 100, color = 'b')
plt.hist(aug, bins = 100, color = 'r')
plt.show()
```
